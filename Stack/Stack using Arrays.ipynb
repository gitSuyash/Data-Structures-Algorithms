# Implement Stack using Arrays

```
# define a stack class & add various operations
class Stack():
  def __init__(self,initial_capacity=10):
    self.arr = [0 for _ in range(initial_capacity)]
    self.next_index = 0
    self.num_elements = 0
```

```
# add push method to the stack
def push(self,value):
    if self.num_elements < len(self.arr):
      self.arr[self.next_index] = value
      self.next_index+=1
      self.num_elements+=1
    else:
      self.handle_full_capacity(value)

Stack.push = push #general way of assigning a method to a class
```

```
# define a method to handle stack overflow situation
def handle_full_capacity(self,value):
    old_arr = self.arr
    self.arr = [0 for _ in range(2*len(self.arr))]
    self.next_index = 0
    self.num_elements=0

    for i, element in enumerate(old_arr):
      self.arr[i] = element
      self.next_index=i+1
      self.num_elements = i+1
    self.arr[self.next_index] = value
    self.next_index+=1
    self.num_elements+=1
Stack.handle_full_capacity = handle_full_capacity
```

```
# define a method to print all elements in a stack
def print_stack(self):
    for i in range(self.num_elements):
      print(self.arr[i])

Stack.print_stack = print_stack
```

```
# define a method to pop an element from head of stack
def pop(self):
    popped = self.arr[self.next_index-1]
    self.arr[self.next_index-1]=0
    self.next_index-=1
    self.num_elements-=1
    return ('element popped',popped)
Stack.pop = pop
```

```
# define a method to obtain size of stack
def size(self):
    print('Number of elements in stack are', self.num_elements)
Stack.size = size
```

```
stack =Stack()
stack.push(1)
stack.push(2)
stack.push(3)
stack.push(4)
stack.push(5)
stack.push(6)
stack.push(7)
stack.push(8)
stack.push(9)
stack.push(10)
stack.push(11)
stack.push(12)

stack.print_stack()
stack.size()
```
